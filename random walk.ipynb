```python
import random
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
jmax = 200  # number of steps
x, y, z = 0.0, 0.0, 0.0 # start at origin

# Store path
x_vals = [x]
y_vals = [y]
z_vals = [z]
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

for step in range(jmax):
    dx = random.uniform(-1,1)
    dy = random.uniform(-1,1)
    dz = random.uniform(-1,1)
    x += dx
    y += dy
    z += dz
    x_vals.append(x)
    y_vals.append(y)
    z_vals.append(z)

    # Update plot
    clear_output(wait=True)
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x_vals, y_vals, z_vals, 'b-', linewidth=1, alpha=0.7, label='path')
    ax.scatter([x_vals[-1]], [y_vals[-1]], [z_vals[-1]], c='r', s=30, label='current')
    ax.scatter([x_vals[0]], [y_vals[0]], [z_vals[0]], c='g', s=50, label='start')
    ax.set_title(f"3D Random Walk (step {step+1}/{jmax})")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.legend()
    plt.show()

print(f"\nFinal position after{jmax} steps: ({x:.2f}, {y:.2f}, {z:.2f})")
```
